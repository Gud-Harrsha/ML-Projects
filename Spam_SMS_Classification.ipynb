## 3) Adding Length and Punctuation columns to dataset

#### Adding length

```python
length = []
for i in range(len(messages)):
    length.append(len(messages['Message'][i]))
```

```python
messages['Length'] = length
messages.head()
```


#### Adding Punctuation

```python
import string
count = 0
punct = []
for i in range(len(messages)):
    for j in messages['Message'][i]:
        if j in string.punctuation:
            count+=1
    punct.append(count)
    count = 0
messages['Punctuation'] = punct
messages.head(5)
```

## 4) Text Cleaning

```python
messages.head(1)
```

```python
import nltk
```

```python
type(nltk.stem)
```

```python
import re, nltk  # Regular Expression Module and Natural Language Tool Kit(2 libraries are imported)

from nltk.corpus import stopwords  # Here nltk.corpus is a Lazy Module. Lazy Module means the import of the module shouldn't 
                                   #occur until the module is actually referenced during execution.
from nltk.stem import WordNetLemmatizer # Here nltk.stem is a module.
# Here stopwords is a module and WordNetLemmatizer is a class.
lemmatizer = WordNetLemmatizer()

corpus = []
#for i in range(2):
for i in range(len(messages)):
        words = re.sub('[^a-zA-Z]', ' ', messages['Message'][i])
        #print(words)
        words = words.lower()
        words = words.split()
        
        #print(words)
        #print(type(words))
        words = [lemmatizer.lemmatize(word) for word in words if word not in stopwords.words('english')]
        #print(words)
        words = ' '.join(words)
        #print(words)
        corpus.append(words)
        #print(corpus)
        #print('--------------------------')
```

```python
type(stopwords.words('english'))
```

#### Printing corpus(The list which contains new messages that are filtered)

```python
corpus[:5]
```

#### Replacing the Message column in messages dataframe with new corpus list

```python
messages['Message'] = corpus
messages.head()
```

#### Recalculating Length and Punctuation of new dataframe after filtering messages

```python
newlength = []
for i in range(len(messages)):
    newlength.append(len(messages['Message'][i]))
messages['NewLength'] = newlength

count = 0
newpunct = []
for i in range(len(messages)):
    for j in messages['Message'][i]:
        if j in string.punctuation:
            count+=1
    newpunct.append(count)
    count = 0
messages['NewPunctuation'] = newpunct
messages.head()
```

## 6) Splitting the 2features from the dataset.

```python
X = messages['Message']
X.head()
```

```python
from sklearn.pipeline import Pipeline
from sklearn.naive_bayes import MultinomialNB
from sklearn.svm import LinearSVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics import classification_report, confusion_matrix
```

```python
#tfidf = TfidfVectorizer()

pipeline_svm = Pipeline([('tfidf', TfidfVectorizer()), ('svm', LinearSVC())])

pipeline_mnb = Pipeline([('tfidf', TfidfVectorizer()), ('mnb', MultinomialNB())])

pipeline_knn = Pipeline([('tfidf', TfidfVectorizer()), ('knn', KNeighborsClassifier())])
```

```python
pipeline_svm[0]
```

## Case 1) test_size = 0.1

```python
from sklearn.model_selection import train_test_split
X_train1, X_test1, y_train1, y_test1 = train_test_split(X, y, test_size = 0.1, train_size = 0.9, random_state=20)
print('length of X_train1 = {} & length of X_test1 = {}'.format(len(X_train1), len(X_test1)))
```

### SVM Classifier

```python
pipeline_svm.fit(X_train1, y_train1)
y_preds_svm1 = pipeline_svm.predict(X_test1)
```

### Naive Bayes Classifier

```python
pipeline_mnb.fit(X_train1, y_train1)
y_preds_mnb1 = pipeline_mnb.predict(X_test1)
```

### KNN Classifier

```python
pipeline_knn.fit(X_train1, y_train1)
y_preds_knn1 = pipeline_knn.predict(X_test1)
```

```python
y_preds1 = [y_preds_svm1, y_preds_mnb1, y_preds_knn1]
classifier = ['SVM CLASSIFIER', 'NAIVEBAYES CLASSIFIER', 'KNN CLASSIFIER']
print('CASE1 : Testsize = 10% and Trainsize = 90%')
for i in range(len(y_preds1)):
    print(classifier[i])
    print(confusion_matrix(y_test1, y_preds1[i]))
    print(classification_report(y_test1, y_preds1[i]))
```
